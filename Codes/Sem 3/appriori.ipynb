```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Generating sample student grades
np.random.seed(42)
grades = np.random.randint(50, 100, 100)  # Creating an array of 100 grades

# Convert to DataFrame
df = pd.DataFrame({'Grades': grades})

# Calculating statistics
mean_grade = df['Grades'].mean()
median_grade = df['Grades'].median()
mode_grade = df['Grades'].mode()

# Creating a histogram
plt.hist(df['Grades'], bins=10)
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.title('Grade Distribution')
plt.show()

print(f"Mean Grade: {mean_grade}")
print(f"Median Grade: {median_grade}")
print(f"Mode Grade: {mode_grade}")
```

```python
from sklearn.linear_model import LinearRegression

# Generating sample data
np.random.seed(42)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 5, 4, 5])  # Dependent variable

# Creating and fitting the regression model
model = LinearRegression()
model.fit(X, y)

# Predicting values
y_pred = model.predict(X)

# Plotting the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.show()
```

```python
# Creating a Pandas DataFrame
data = {
    'A': np.random.randint(1, 100, 5),
    'B': np.random.randint(1, 100, 5),
    'C': np.random.randint(1, 100, 5)
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Binning
bins = [0, 30, 60, 100]
labels = ['Low', 'Medium', 'High']
df['Category'] = pd.cut(df['A'], bins=bins, labels=labels)
print("\nBinned DataFrame:")
print(df)
```

```python
# Creating a DataFrame with hierarchical indexing
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': ['X', 'Y', 'X', 'Y', 'Z']
}
arrays = [['Group1', 'Group1', 'Group2', 'Group2', 'Group3'], [1, 2, 1, 2, 1]]
index = pd.MultiIndex.from_arrays(arrays, names=('Group', 'Subgroup'))
df = pd.DataFrame(data, index=index)
print("DataFrame with Hierarchical Index:")
print(df)

# Selecting data based on hierarchical index
print("\nSelecting data from 'Group2':")
print(df.loc['Group2']) 
```

```python
# Creating a sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Values': [10, 20, 15, 25, 12, 18]
}
df = pd.DataFrame(data)

# Grouping by 'Category' and computing mean and sum for each group
grouped = df.groupby('Category').agg({'Values': ['mean', 'sum']})

print("Grouped DataFrame with Aggregated Values:")
print(grouped)
```

```python
# Creating a sample DataFrame
np.random.seed(0)
x = np.random.normal(-3, 4, 1000)
y = np.random.normal(-3, 4, 1000)
df = pd.DataFrame({'X': x, 'Y': y})

# Creating a hexbin plot
df.plot(kind='hexbin', x='X', y='Y', gridsize=20, cmap='coolwarm', figsize=(8, 6))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Hexbin Plot')
plt.tight_layout()
plt.show()
```
