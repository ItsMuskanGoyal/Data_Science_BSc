2. Do the following using PANDAS Series:
a. Create a series with 5 elements. Display the series sorted on index and also sorted on values seperately
b. Create a series with N elements with some duplicate values. Find the minimum and maximum ranks
assigned to the values using ‘first’ and ‘max’ methods
c. Display the index value of the minimum and maximum element of a Series

```python
import pandas as pd

# Creating a series with 5 elements
series_a = pd.Series([4, 2, 7, 1, 5], index=['e', 'b', 'd', 'a', 'c'])

# Sorting the series by index and values separately
sorted_by_index = series_a.sort_index()
sorted_by_values = series_a.sort_values()

print("Sorted by index:\n", sorted_by_index)
print("\nSorted by values:\n", sorted_by_values)




# Creating a series with duplicate values
series_b = pd.Series([10, 5, 7, 5, 2, 10, 7])

# Finding minimum and maximum ranks using 'first' and 'max' methods
min_ranks_first = series_b.rank(method='first', ascending=True)
max_ranks_max = series_b.rank(method='max', ascending=True)

min_rank_value = min_ranks_first.min()
max_rank_value = max_ranks_max.max()

print("Minimum ranks assigned using 'first' method:\n", min_ranks_first)
print("\nMaximum ranks assigned using 'max' method:\n", max_ranks_max)
print("\nMinimum rank value:", min_rank_value)
print("Maximum rank value:", max_rank_value)




# Finding index value of minimum and maximum elements
min_index = series_b.idxmin()
max_index = series_b.idxmax()

print("Index of the minimum element:", min_index)
print("Index of the maximum element:", max_index)
```

3. Create a data frame having at least 3 columns and 50 rows to store numeric data generated using a random
function. Replace 10% of the values by null values whose index positions are generated using random function.
Do the following:
a. Identify and count missing values in a data frame.
b. Drop the column having more than 5 null values.
c. Identify the row label having maximum of the sum of all values in a row and drop that row.
d. Sort the data frame on the basis of the first column.
e. Remove all duplicates from the first column.
f. Find the correlation between first and second column and covariance between second and third
column.
g. Discretize the second column and create 5 bins.

```python
import pandas as pd
import numpy as np

# Creating a DataFrame with random numeric data
np.random.seed(42)  # Setting seed for reproducibility
data = np.random.rand(50, 3)  # Generating random data for 50 rows and 3 columns
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

# Replacing 10% of values with NaN
nan_indices = np.random.choice(df.size, replace=False, size=int(0.1 * df.size))
df.values.ravel()[nan_indices] = np.nan

# Identifying and counting missing values
missing_values_count = df.isnull().sum()
print("Missing values count:\n", missing_values_count)




# Dropping columns with more than 5 null values
df_dropped = df.dropna(thresh=df.shape[0]-5, axis=1)
print("\nDataFrame after dropping columns:\n", df_dropped)




# Identifying row label with the maximum sum and dropping that row
max_sum_row_label = df.sum(axis=1).idxmax()
df_dropped_row = df.drop(index=max_sum_row_label)
print("\nDataFrame after dropping row with maximum sum:\n", df_dropped_row)
```

```python
# Sorting the DataFrame based on the first column
df_sorted = df.sort_values(by='A')
print("\nDataFrame sorted by the first column:\n", df_sorted)




# Removing duplicates from the first column
df_no_duplicates = df.drop_duplicates(subset='A')
print("\nDataFrame after removing duplicates from the first column:\n", df_no_duplicates)




# Finding correlation and covariance
correlation_first_second = df['A'].corr(df['B'])
covariance_second_third = df['B'].cov(df['C'])

print("\nCorrelation between first and second column:", correlation_first_second)
print("Covariance between second and third column:", covariance_second_third)




# Discretizing the second column into 5 bins
# df['B_discretized'] = pd.cut(df['B'], bins=5)
# print("\nDataFrame after discretizing the second column into 5 bins:\n", df)
```

4. Consider two excel files having attendance of two workshops, each of duration 5 days. Each file has three
fields ‘Name’, ‘Date, duration (in minutes) where names may be repetitve within a file. Note that duration may
take one of three values (30, 40, 50) only. Import the data into two data frames and do the following:
a. Perform merging of the two data frames to find the names of students who had attended both
workshops.
b. Find names of all students who have attended a single workshop only.

c. Merge two data frames row-wise and find the total number of records in the data frame.
d. Merge two data frames row-wise and use two columns viz. names and dates as multi-row indexes.
Generate descriptive statistics for this hierarchical data frame.
Adjust the file paths ('workshop1_attendance.xlsx', 'workshop2_attendance.xlsx') to the actual paths of your Excel files. These snippets read the Excel files into dataframes, perform merging, find attendees, merge row-wise, and compute descriptive statistics for hierarchical dataframes based on the workshop attendance data.

```python
import pandas as pd

# Reading the Excel files into dataframes
workshop1_df = pd.read_excel('att1.xlsx')
workshop2_df = pd.read_excel('att2.xlsx')

# Performing merging to find names attending both workshops
both_attended = pd.merge(workshop1_df, workshop2_df, on='Name')
names_both_workshops = both_attended['Name'].unique()
print("Names of students who attended both workshops:\n", names_both_workshops)




# Finding names of students attending a single workshop only
single_workshop1 = workshop1_df[~workshop1_df['Name'].isin(names_both_workshops)]
single_workshop2 = workshop2_df[~workshop2_df['Name'].isin(names_both_workshops)]

names_single_workshop = pd.concat([single_workshop1['Name'], single_workshop2['Name']]).unique()
print("\nNames of students who attended a single workshop only:\n", names_single_workshop)




# Merging dataframes row-wise and finding the total number of records
merged_row_wise = pd.concat([workshop1_df, workshop2_df])
total_records = merged_row_wise.shape[0]
print("\nTotal number of records after merging row-wise:", total_records)




# Merging dataframes row-wise and setting multi-row indexes
merged_row_wise_multi_index = pd.concat([workshop1_df, workshop2_df], keys=['Workshop1', 'Workshop2'])

# Generating descriptive statistics for the hierarchical dataframe
descriptive_stats = merged_row_wise_multi_index.groupby(level=0).describe()
print("\nDescriptive statistics for hierarchical dataframe:\n", descriptive_stats)
```

6. Using Titanic dataset, to do the following:
a. Clean the data by dropping the column which has the largest number of missing values.
b. Find total number of passengers with age more than 30
c. Find total fare paid by passengers of second class
d. Compare number of survivors of each passenger class
e. Compute descriptive statistics for age attribute gender wise
f. Draw a scatter plot for passenger fare paid by Female and Male passengers separately
g. Compare density distribution for features age and passenger fare
h. Draw the pie chart for three groups labelled as class 1, class 2, class 3 respectively displayed in different
colours. The occurrence of each group converted into percentage should be displayed in the pie chart.
Appropriately Label the chart.
i. Find % of survived passengers for each class and answer the question “Did class play a role in survival?”.

```python
import pandas as pd

# Load Titanic dataset
titanic_data = pd.read_csv('train.csv')

# Find column with the largest number of missing values
max_missing_col = titanic_data.isnull().sum().idxmax()

# Drop the column with the largest number of missing values
cleaned_data = titanic_data.drop(columns=[max_missing_col])




# Total number of passengers with age more than 30
passengers_over_30 = cleaned_data[cleaned_data['Age'] > 30]
total_passengers_over_30 = passengers_over_30.shape[0]
print("Total passengers with age more than 30:", total_passengers_over_30)




# Total fare paid by passengers of second class
total_fare_second_class = cleaned_data[cleaned_data['Pclass'] == 2]['Fare'].sum()
print("Total fare paid by passengers of second class:", total_fare_second_class)




# Compare number of survivors of each passenger class
survivors_per_class = cleaned_data.groupby('Pclass')['Survived'].sum()
print("Number of survivors of each passenger class:\n", survivors_per_class)
```

```python
# Compute descriptive statistics for age attribute gender wise
descriptive_stats_age_gender = cleaned_data.groupby('Sex')['Age'].describe()
print("Descriptive statistics for age attribute gender wise:\n", descriptive_stats_age_gender)



import scipy as sp
import matplotlib.pyplot as plt

# Scatter plot for passenger fare paid by Female and Male passengers separately
plt.scatter(cleaned_data[cleaned_data['Sex'] == 'female']['Fare'], cleaned_data[cleaned_data['Sex'] == 'female']['Sex'], label='Female')
plt.scatter(cleaned_data[cleaned_data['Sex'] == 'male']['Fare'], cleaned_data[cleaned_data['Sex'] == 'male']['Sex'], label='Male')
plt.xlabel('Fare')
plt.ylabel('Sex')
plt.legend()
plt.title('Passenger Fare Paid by Female and Male Passengers')
plt.show()




# Density distribution for features age and passenger fare
cleaned_data[['Age', 'Fare']].plot(kind='density', subplots=True, layout=(1, 2))
plt.tight_layout()
plt.show()




# Draw a pie chart for three passenger classes
class_counts = cleaned_data['Pclass'].value_counts()
plt.pie(class_counts, labels=['Class 1', 'Class 2', 'Class 3'], autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Passenger Class Distribution')
plt.show()




# % of survived passengers for each class
survival_percent_per_class = cleaned_data.groupby('Pclass')['Survived'].mean() * 100
print("Percentage of survived passengers for each class:\n", survival_percent_per_class)
```

Write a program in Python using Pandas to perform the following:
a. Calculate and display familywise gross monthly income.

b. Display the highest and lowest monthly income for each family name
c. Calculate and display monthly income of all members earning income less than Rs. 80000.00.
d. Display total number of females along with their average monthly income
e. Delete rows with Monthly income less than the average income of all members

```python
import pandas as pd

# Creating the DataFrame
data = {
    'FamilyName': ['Shah', 'Vats', 'Vats', 'Kumar', 'Vats', 'Kumar', 'Shah', 'Shah', 'Kumar', 'Vats'],
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male'],
    'MonthlyIncome': [44000.00, 65000.00, 43150.00, 66500.00, 255000.00, 103000.00, 55000.00, 112400.00, 81030.00, 71900.00]
}
df = pd.DataFrame(data)

# Calculate and display familywise gross monthly income
familywise_income = df.groupby('FamilyName')['MonthlyIncome'].sum()
print("Familywise gross monthly income:\n", familywise_income)




# Display the highest and lowest monthly income for each family name
highest_income = df.groupby('FamilyName')['MonthlyIncome'].max()
lowest_income = df.groupby('FamilyName')['MonthlyIncome'].min()
print("\nHighest monthly income for each family name:\n", highest_income)
print("\nLowest monthly income for each family name:\n", lowest_income)




# Calculate and display monthly income of all members earning income less than Rs. 80000.00
income_less_than_80000 = df[df['MonthlyIncome'] < 80000.00]['MonthlyIncome']
print("\nMonthly income of members earning less than Rs. 80000.00:\n", income_less_than_80000)




# Display total number of females along with their average monthly income
females = df[df['Gender'] == 'Female']
total_females = females.shape[0]
average_income_females = females['MonthlyIncome'].mean()
print("\nTotal number of females:", total_females)
print("Average monthly income of females:", average_income_females)




# Delete rows with Monthly income less than the average income of all members
average_income_all = df['MonthlyIncome'].mean()
df = df[df['MonthlyIncome'] >= average_income_all]
print("\nDataFrame after deleting rows with income less than the average:\n", df)
```
