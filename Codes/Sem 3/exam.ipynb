```python

x_val = np.arange(0, 2*np.pi,0.1)
plt.figure(figsize = (6,6))

plt.subplot(2,2,1)
y_cos = np.cos(x_val)
plt.plot(x_val, y_cos, 'r--')

plt.subplot(2,2,2)
y_sin = np.sin(x_val)
plt.plot(x_val, y_sin, 'b--')


plt.subplot(2,2,3)
y_tan = np.tan(x_val)
plt.plot(x_val, y_tan, 'g--')

plt.subplot(2,2,4)
y_cos = np.cos(x_val)
plt.plot(x_val, y_cos, 'w--')

plt.show()


```
